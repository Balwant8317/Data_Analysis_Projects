# Data Analysis And Visulization

```python
## Here we are checking how much reservations are cancelled and not cancelled in the from of percentage

cancelled_perc = df["is_canceled"].value_counts(normalize=True)
```

```python
cancelled_perc
```

Here we observed around 62% reservation are not cancelled bout 32% are cancelled

```python
### Here we are visulizing canceled and not canceled reservation

plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(["Not Canceled","Canceled"],df["is_canceled"].value_counts(), edgecolor = 'r',width = 0.7)
plt.show()
```

```python
## Kiska Cancelation ration zayda hai

plt.figure(figsize=(5,4))
ax1 = sns.countplot(x = 'hotel',hue='is_canceled',data = df, palette = "Blues")
legend_labeles,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))

plt.title("Reservation status in differnt hotels",size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
### How many percent reservation cancel and not cancel basically here we finding percentage in resort hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']

resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
### How many percent reservation cancel and not cancel basically here we finding percentage in city hotel
city_hotel = df[df['hotel'] == 'City Hotel']

city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
## Is there any price effect in city hotel and resort hotel due to reservation cancellation

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
## create average daily rate for city hotel and resort hotel

plt.figure(figsize=(20,8))

plt.title("Average Daily rate in city and resort hotel",fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label= 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.show()
```

```python
## here we checking how much reservation in month and how much cancle reservation in month
### Here also i am using countplot
df['month'] = df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month',hue = 'is_canceled',data=df)
```

```python
##Avarge daily rate for each month
plt.figure(figsize=(16,8))
plt.title("ADR per month", fontsize = 30)

sns.barplot('month','adr',data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.legend(fontsize=20)
plt.show
```

```python
#### we are checking cancellation rate on the basis of country we are checking only top ten countries

cancelled_data = df[df['is_canceled']==1]

top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("top ten countrys with reservation cancaled")
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
```
